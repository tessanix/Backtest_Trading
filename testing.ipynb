```python
import pandas as pd

def create_df(start_date="2023-03-24 12:00:00", end_date="2025-02-14 12:00:00" ):
    filename = "C:/Users/tessa/Codes/Backtest_Trading/market_data/market_news.csv"
        
    df = pd.read_csv(
        filepath_or_buffer=filename,
        delimiter=",",
    )
    df = df[(df['currency']=="USD") & (df['impact'] == "red")]
    df['datetime'] = df['date'] + ' ' + df['time']
    df['datetime'] = df['datetime'].replace('2024 Nov 5 All Day', '2024 Nov 5 2:30pm', regex=True)
    df['datetime'] = pd.to_datetime(df['datetime'], format='%Y %b %d %I:%M%p')
    #df = df[( pd.to_datetime(start_date) <= df['datetime']) & (df['datetime'] <=  pd.to_datetime(end_date))]
    df.drop(axis=1, labels=["time", "date"], inplace=True)
    df.reset_index(inplace=True, drop=True)
    return df[["datetime", "currency", "impact", "event"]]
```

```python
df = create_df()
df
```
